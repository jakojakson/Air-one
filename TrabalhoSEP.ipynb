##Valor das Impedancias

```python
#Impedancias PU

def imped(impedancia_real,distancia,quantia,zbase):
  resultado = (impedancia_real*distancia/quantia)/zbase
  return resultado

'''
Z1= 0.3703 + 1.3690j
Zcmt1= 0.0169 + 0.000206j
Zcmt2= 0.0282109850871666 + 0.006545j
Zcbt1= 12.170 + 4.1892j
Zcbt2= 0.721675900277008 + 0.00091205j
Zcqda= 114.85 + 14.91j
Zcqdb= 92.045 + 12.0247j
Zcqdc= 36.77 + 9.31j
Zcqdd= 3.04709141274238 + 0.0025j
Zcqde= 2.63157894736842 + 0.0027j

'''
Z1= 0.3703 + 1.3690j
Zcmt1= imped(1.535 + 0.187j,21e-3,1,Zbmt)
print("Zcmt1 = ", Zcmt1)
Zcmt2= imped(1.535 + 0.187j,35e-3,1,Zbmt)
print("Zcmt2 = ", Zcmt2)
Zcbt1= imped(0.3184 + 0.1096j,37e-3,2,Zbbt1)
print("Zcbt1 = ", Zcbt1)
Zcbt2= imped(0.1226 + 0.1073j,34e-3,4,Zbbt2)
print("Zcbt2 = ", Zcbt2)
Zcqda= imped(0.8891 + 0.1164j,62e-3,1,Zbbt1)
print("Zcqda = ", Zcqda)
Zcqdb= imped(0.8891 + 0.1164j,50e-3,1,Zbbt1)
print("Zcqdb = ", Zcqdb)
Zcqdc= imped(0.4450 + 0.1127j,40e-3,1,Zbbt1)
print("Zcqdc = ", Zcqdc)
Zcqdd= imped(0.1868 + 0.1076j,70e-3,3,Zbbt2)
print("Zcqdd = ", Zcqdd)
Zcqde= imped(0.1502 + 0.1074j,25e-3,1,Zbbt2)
print("Zcqde = ", Zcqde)


#Transformador 1
SnT1=150e3
ZT1=3.5
PcobreT1=545
RTpu1=PcobreT1/SnT1
XlTpu1=np.sqrt(((ZT1/100)**2)-RTpu1**2)
Ztr1=(RTpu1+XlTpu1*1j)*Sb/SnT1
print("Ztr1 = ", Ztr1)

#Transformador 2
SnT2=1000e3
ZT2=6
PcobreT2=10600
RTpu2=PcobreT2/SnT2
XlTpu2=np.sqrt(((ZT2/100)**2)-RTpu2**2)
Ztr2=(RTpu2+XlTpu2*1j)*Sb/SnT2
print("Ztr2 = ", Ztr2)



```

##Valor de ReferÃªncia

```python
#FunÃ§Ãµes auxiliares 

def imaginario(valor_real, fp):   #FunÃ§Ã£o para cÃ¡lculo de potÃªncia em valor complexo conjugado
  rad = np.arccos(fp)
  Sx = rect(valor_real,rad)
  return Sx


#Valores das cargas
Valor_real_QDA = 51e3
fpQDA = 0.87
S1 = imaginario(Valor_real_QDA,fpQDA)*1
print("S1 = ",S1)

Valor_real_QDB = 38e3
fpQDB = 0.89
S2 = imaginario(Valor_real_QDB,fpQDB)*1
print("S2 = ", S2)

Valor_real_QDC = -60e3
fpQDC = 1
S3 = imaginario(Valor_real_QDC,fpQDC)*1
print("S3 = ", S3)

Valor_real_QDD = 601e3
fpQDD = 0.85
S4 = imaginario(Valor_real_QDD,fpQDD)*1
print("S4 = ", S4)

Valor_real_QDE = 228e3
fpQDE = 0.87
S5 = imaginario(Valor_real_QDE,fpQDE)*1
print("S5 = ", S5)

S6 = complex(0 - 20e3j)*1
print("S6 = ",S6)

S7 = (0 - 85e3j)*1
print("S7 = ",S7)


#ConversÃ£o das cargas para PU
S1pu=S1/Sb
S2pu=S2/Sb
S3pu=S3/Sb
S4pu=S4/Sb
S5pu=S5/Sb
S6pu=S6/Sb
S7pu=S7/Sb

#DeclaraÃ§Ã£o das condiÃ§Ãµes iniciais em PU (todas as tensÃµes em 1pu - cargas desligadas)

Vqdepu = 1+0j     #TensÃ£o no QD-E
Vqddpu = 1+0j     #TensÃ£o no QD-D
Vqdcpu = 1+0j     #TensÃ£o no QD-C
Vqdbpu = 1+0j     #TensÃ£o no QD-B
Vqdapu = 1+0j     #TensÃ£o no QD-A
Vqdgpu2 = 1+0j    #TensÃ£o no QDG 2
Vqdgpu1 = 1+0j    #TensÃ£o no QDG 1
VTbtpu2 = 1+0j    #TensÃ£o no Lado de Baixa TensÃ£o do Trafo2
VTbtpu1 = 1+0j    #TensÃ£o no Lado de Baixa TensÃ£o do Trafo1
VTmtpu2 = 1+0j    #TensÃ£o no Lado de MÃ©dia TensÃ£o do Trafo2
VTmtpu1 = 1+0j    #TensÃ£o no Lado de MÃ©dia TensÃ£o do Trafo1
VPepu = 1+0j      #TensÃ£o no Ponto de Entrega
VSEpu = 1+0j      #TensÃ£o na SubestaÃ§Ã£o da concessionÃ¡ria - ReferÃªncia (sempre em 1PU)
```

##Valores por mÃ©todo de Gauss

```python
for x in range(200):
  #CÃ¡lculo das correntes

  ISqdapu = S1pu/Vqdgpu1
  ISqdapu = ISqdapu.conjugate()

  ISqdbpu = S2pu/Vqdgpu1
  ISqdbpu = ISqdbpu.conjugate()

  ISqdcpu = S3pu/Vqdapu
  ISqdcpu = ISqdcpu.conjugate()

  ISqddpu = S4pu/Vqdgpu2
  ISqddpu = ISqddpu.conjugate()

  ISqdepu = S5pu/Vqdgpu2
  ISqdepu = ISqdepu.conjugate()

  ISqd1pu = S6pu/Vqdgpu1
  ISqd1pu = ISqd1pu.conjugate()

  ISqd2pu = S7pu/Vqdgpu2
  ISqd2pu = ISqd2pu.conjugate()

  ISqdapu = ISqdapu + ISqdcpu

  Iqdpu1 = ISqdapu + ISqdbpu  + ISqd1pu

  Iqdpu2 = ISqddpu + ISqdepu + ISqd2pu 

  Itpu = Iqdpu1 + Iqdpu2 

 
  #CÃ¡lculo das quedas de tensÃµes
  DVZQDE = ISqdepu * Zcqde              #Cabo do QDE
  DVZQDD = ISqddpu * Zcqdd              #Cabo do QDD
  DVZQDC = ISqdcpu * Zcqdc              #Cabo do QDC
  DVZQDB = ISqdbpu * Zcqdb              #Cabo do QDB
  DVZQDA = ISqdapu * Zcqda   #Cabo do QDA 
  DVZcbt2 = Iqdpu2*Zcbt2     #Cabo de BT 2
  DVZcbt1 = Iqdpu1*Zcbt1     #Cabo de BT 1
  DVtr2 = Iqdpu2*Ztr2       #Transformador 2
  DVtr1 = Iqdpu1*Ztr1       #Transformador 1
  DVZcmt2 = Iqdpu2*Zcmt2    #Cabo de MT 2
  DVZcmt1 = Iqdpu1*Zcmt1    #Cabo de MT 1
  DVZ1 = Itpu*Z1            #ImpedÃ¢ncia da rede (thevenin)

  #CÃ¡lculo das tensÃµes nos pontos de inreresse - em PU


  VPepu = VSEpu-DVZ1
  VTmtpu1 = VPepu-DVZcmt1
  VTmtpu2 = VPepu-DVZcmt2
  VTbtpu1 = VTmtpu1-DVtr1
  VTbtpu2 = VTmtpu2-DVtr2
  Vqdgpu1 = VTbtpu1-DVZcbt1
  Vqdgpu2 = VTbtpu2-DVZcbt2
  Vqdapu = Vqdgpu1-DVZQDA
  Vqdbpu = Vqdgpu1-DVZQDB
  Vqdcpu = Vqdapu-DVZQDC
  Vqddpu = Vqdgpu2-DVZQDD
  Vqdepu = Vqdgpu2-DVZQDE
```

##CÃ¡culo de valores reais

```python
#TensÃµes reais
VPe=VPepu*Vbmt

VTrmt1=VTmtpu1*Vbmt
VTrmt2=VTmtpu2*Vbmt

VTrbt1=VTbtpu1*Vbbt1
VTrbt2=VTbtpu2*Vbbt2

Vqdg1=Vqdgpu1*Vbbt1
Vqdg2=Vqdgpu2*Vbbt2

Vqda = Vqdapu*Vbbt1
Vqdb = Vqdbpu*Vbbt1
Vqdc = Vqdcpu*Vbbt1

Vqdd = Vqddpu*Vbbt2
Vqde = Vqdepu*Vbbt2

#Correntes reais

ISQDE = ISqdepu*Ibbt2 #Corrente real no QD-E
ISQDD = ISqddpu*Ibbt2 #Corrente real no QD-D

ISQDC = ISqdcpu*Ibbt1 #Corrente real no QD-C
ISQDB = ISqdbpu*Ibbt1 #Corrente real no QD-B
ISQDA = ISqdapu*Ibbt1 #Corrente real no QD-A

IQDG2 = Iqdpu2*Ibbt2  #Corrente real no QDG 2
IQDG1 = Iqdpu1*Ibbt1  #Corrente real no QDG 1

IMt2 = Iqdpu2 * Ibmt  #Corrente real no lado de mÃ©dia tensÃ£o do transformador 2
IMt1 = Iqdpu1 * Ibmt  #Corrente real no lado de mÃ©dia tensÃ£o do transformador 1

IPe=Itpu*Ibmt         #Corrente real no ponto de entrega


#PotÃªncias reais
SQDE = (Vqdepu*ISqdepu.conjugate())*Sb    #PotÃªncia real no QD-E
SQDD = (Vqddpu*ISqddpu.conjugate())*Sb    #PotÃªncia real no QD-D
SQDC = (Vqdcpu*ISqdcpu.conjugate())*Sb    #PotÃªncia real no QD-C
SQDB = (Vqdbpu*ISqdbpu.conjugate())*Sb    #PotÃªncia real no QD-B
SQDA = (Vqdapu*ISqdapu.conjugate())*Sb    #PotÃªncia real no QD-A

SQDG2= (Vqdgpu2*Iqdpu2.conjugate())*Sb    #PotÃªncia real no QDG 2
SQDG1= (Vqdgpu1*Iqdpu1.conjugate())*Sb    #PotÃªncia real no QDG 1

STbt2= (VTbtpu2*Iqdpu2.conjugate())*Sb   #PontÃªncia real no lado de baixa tensÃ£o do transformador 2
STbt1= (VTbtpu1*Iqdpu1.conjugate())*Sb   #PontÃªncia real no lado de baixa tensÃ£o do transformador 1

SMt2 = (VTmtpu2*Iqdpu2.conjugate())*Sb   #PontÃªncia real no lado de mÃ©dia tensÃ£o do transformador 2
SMt1 = (VTmtpu1*Iqdpu1.conjugate())*Sb   #PontÃªncia real no lado de mÃ©dia tensÃ£o do transformador 1

SPe=(VPepu*Itpu.conjugate())*Sb           #PontÃªncia real no ponto de entrega
```

#VisualizaÃ§Ã£o dos Dados

```python
Tensao= [abs(VPe), abs(VTrmt1), abs(VTrmt2) ,abs(VTrbt1), abs(VTrbt2), abs(Vqdg1), abs(Vqdg2), abs(Vqda), abs(Vqdb), abs(Vqdc), abs(Vqdd), abs(Vqde)]

Angulo_V= [phase(VPe)*(180/pi), phase(VTrmt1)*(180/pi), phase(VTrmt2)*(180/pi), phase(VTrbt1)*(180/pi), phase(VTrbt2)*(180/pi), phase(Vqdg1)*(180/pi), phase(Vqdg2)*(180/pi), phase(Vqda)*(180/pi), phase(Vqdb)*(180/pi), phase(Vqdc)*(180/pi), phase(Vqdd)*(180/pi), phase(Vqde)*(180/pi)]

Corrente=[abs(IPe), abs(IMt1), abs(IMt2), abs(IQDG1), abs(IQDG2), abs(IQDG1), abs(IQDG2), abs(ISQDA), abs(ISQDB), abs(ISQDC), abs(ISQDD), abs(ISQDE)]

Angulo_I= [phase(IPe)*(180/pi), phase(IMt1)*(180/pi), phase(IMt2)*(180/pi), phase(IQDG1)*(180/pi), phase(IQDG2)*(180/pi), phase(IQDG1)*(180/pi), phase(IQDG2)*(180/pi), phase(ISQDA)*(180/pi), phase(ISQDB)*(180/pi), phase(ISQDC)*(180/pi), phase(ISQDD)*(180/pi), phase(ISQDE)*(180/pi)]

S_kVA= [abs(SPe)*1/1000, abs(SMt1)*1/1000, abs(SMt2)*1/1000, abs(STbt1)*1/1000, abs(STbt2)*1/1000, abs(SQDG1)*1/1000, abs(SQDG2)*1/1000, abs(SQDA)*1/1000, abs(SQDB)*1/1000, abs(SQDC)*1/1000, abs(SQDD)*1/1000, abs(SQDE)*1/1000]

P_kW=[SPe.real*1/1000, SMt1.real*1/1000, SMt2.real*1/1000, STbt1.real*1/1000, STbt2.real*1/1000, SQDG1.real*1/1000, SQDG2.real*1/1000, SQDA.real*1/1000, SQDB.real*1/1000, SQDC.real*1/1000, SQDD.real*1/1000, SQDE.real*1/1000]

FP=[np.cos(phase(SPe)), np.cos(phase(SMt1)), np.cos(phase(SMt2)), np.cos(phase(STbt1)), np.cos(phase(STbt2)), np.cos(phase(SQDG1)), np.cos(phase(SQDG2)), np.cos(phase(SQDA)), np.cos(phase(SQDB)), np.cos(phase(SQDC)), np.cos(phase(SQDD)), np.cos(phase(SQDE))]

Dados = [[Tensao],[Angulo_V],[Corrente],[Angulo_I],[S_kVA],[P_kW],[FP]]

prop_rates = pd.DataFrame([Tensao, Angulo_V, Corrente, Angulo_I, S_kVA, P_kW , FP], columns=['Ponto Entrega','PrimÃ¡rio Tr1', 'PrimÃ¡rio Tr2', 'SecundÃ¡rio Tr1', 'SecundÃ¡rio Tr2','QDG1','QDG2','QD-A','QD-B','QD-C','QD-D','QD-E'], index=['TensÃ£o','Ãngulo_V','Corrente','Angulo_I','S[kVA]','P[kW]','FP']).transpose()

prop_rates.round(2)

```
